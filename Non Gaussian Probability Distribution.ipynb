# QQ Plot to prob normal Distribution

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset("iris")
df
```

```python
sns.kdeplot(df["sepal_length"])
```

```python
temp=sorted((df["sepal_length"]).tolist())
```

```python
x_quant=[]
for i in range(1, 101):
    x_quant.append(np.percentile(temp, i))
```

```python
x_quant
```

```python
samples=np.random.normal(loc=0, scale=1, size=1000)
```

```python
y_quant=[]

for i in range(1, 101):
    y_quant.append(np.percentile(samples, i))
```

```python
y_quant
```

```python
sns.scatterplot(x=x_quant, y=y_quant)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
fig=sm.qqplot(df["sepal_length"], line="45", fit=True)
plt.title("QQ plot")
plt.show()
```

```python
x=np.random.uniform(low=0, high=1, size=10000)
```

```python
plt.hist(x)
```

```python
from scipy import stats
params=stats.uniform.fit(x)
```

```python
dist=stats.uniform(loc=params[0], scale=params[1])
```

```python
sm.qqplot(x, dist=dist, line="45")
plt.title("QQ plot of Uniform Distribution with Uniform fit")
plt.show()
```

# log log plot for Pareto distribution

```python
alpha=3
xm=1

x=np.linspace(0.1, 10, 1000)
```

```python
y=alpha*(xm**alpha)/x**(alpha+1)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(np.log(x), np.log(y))
plt.title("Pareto Distribution")
plt.xlabel("Log x")
plt.ylabel("Log y")
plt.show()
```

```python
x=stats.pareto.rvs(b=alpha, scale=xm, size=1000)
```

```python
plt.hist(x)
plt.show()
```

```python
params=stats.pareto.fit(x, floc=0)
```

```python
dist=stats.pareto(b=params[0], scale=params[2])
```

```python
fig=sm.qqplot(x, dist=dist, line="45")
plt.title("QQplot of Pareto Distribution fit")
plt.show()
```

# Transormers - Function Transformer - Log Transformer

```python
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv("titanic.csv")[["Age", "Fare", "Survived"]]
df
```

```python
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:, 0:2]
y=df.iloc[:,2]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
sns.distplot(X_train["Age"])
plt.title("Age PDF")

plt.subplot(122)
stats.probplot(X_train["Age"], dist="norm", plot=plt)
plt.title("Age QQ plot")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
sns.distplot(X_train["Fare"])
plt.title("Fare PDF")

plt.subplot(122)
stats.probplot(X_train["Fare"], dist="norm", plot=plt)
plt.title("Fare QQ plot")
plt.show()
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
```

```python
lr.fit(X_train, y_train)
```

```python
dt.fit(X_train, y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
dt_pred=dt.predict(X_test)
```

```python
print(f"LR accuracy : {accuracy_score(lr_pred, y_test)}")
print(f"DT accuracy : {accuracy_score(dt_pred, y_test)}")
```

```python
trf=FunctionTransformer(func=np.log1p)
```

```python
X_train_trf=trf.fit_transform(X_train)
X_test_trf=trf.transform(X_test)
```

```python
lr.fit(X_train_trf, y_train)
```

```python
dt.fit(X_train_trf, y_train)
```

```python
lr_pred_trf=lr.predict(X_test_trf)
```

```python
dt_pred_trf=dt.predict(X_test_trf)
```

```python
print(f"LR Trf accuracy : {accuracy_score(lr_pred_trf, y_test)}")
print(f"DT Trf accuracy : {accuracy_score(dt_pred_trf, y_test)}")
```

```python
X_transformed=trf.fit_transform(X)
```

```python
print("LR Cross val score :", np.mean(cross_val_score(lr, X_transformed, y, scoring="accuracy", cv=10)))
print("DT Cross val score :", np.mean(cross_val_score(dt, X_transformed, y, scoring="accuracy", cv=10)))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
stats.probplot(X_train["Age"], dist="norm", plot=plt)
plt.title("Age before Log")

plt.subplot(122)
stats.probplot(X_train_trf["Age"], dist="norm", plot=plt)
plt.title("Age after Log")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
stats.probplot(X_train["Fare"], dist="norm", plot=plt)
plt.title("Fare before Log")

plt.subplot(122)
stats.probplot(X_train_trf["Fare"], dist="norm", plot=plt)
plt.title("Fare after Log")
plt.show()
```

```python
final_trf=ColumnTransformer([("log", FunctionTransformer(np.log1p), ["Fare"])], remainder="passthrough")
```

```python
X_train_final_trf=final_trf.fit_transform(X_train)
X_test_final_trf=final_trf.transform(X_test)
```

```python
lr.fit(X_train_final_trf, y_train)
dt.fit(X_train_final_trf, y_train)
```

```python
lr_pred_final=lr.predict(X_test_final_trf)
dt_pred_final=dt.predict(X_test_final_trf)
```

```python
print("Accuracy for lr :", accuracy_score(lr_pred_final, y_test))
print("Accuracy for dt :", accuracy_score(dt_pred_final, y_test))
```
