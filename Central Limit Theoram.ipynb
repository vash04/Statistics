# Central Limit Theoeram in Code

```python
num_samples=1000
sample_size=30
distribution_range=(0,1)

samples=np.random.uniform(distribution_range[0], distribution_range[1], (num_samples, sample_size))

sample_means=np.mean(samples, axis=1)

plt.hist(sample_means, bins=30, density=True, edgecolor="black")
plt.title("Histogram of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
num_samples=1000
sample_size=30

poisson_lambda=5

gamma_shape=2
gamma_scale=1

binomial_n=10
binomial_p=0.5

poisson_samples=np.random.poisson(lam=poisson_lambda, size=(num_samples, sample_size))
gamma_samples=np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
binomial_samples=np.random.binomial(n=binomial_n, p=binomial_p, size=(num_samples, sample_size))

poisson_sample_means=np.mean(poisson_samples, axis=1)
gamma_sample_means=np.mean(gamma_samples, axis=1)
binomial_sample_means=np.mean(binomial_samples, axis=1)

fig, axs=plt.subplots(3, 1, figsize=(8,15))

axs[0].hist(poisson_sample_means, bins=30, density=True, edgecolor="black")
axs[0].set_title("Histogram of Sample Means for Poisson Distribution")
axs[0].set_xlabel("Sample Mean")
axs[0].set_ylabel("Density")

axs[1].hist(gamma_sample_means, bins=30, density=True, edgecolor="black")
axs[1].set_title("Histogram of Sample Means for Gamma Distribution")
axs[1].set_xlabel("Sample Mean")
axs[1].set_ylabel("Density")

axs[2].hist(binomial_sample_means, bins=30, density=True, edgecolor="black")
axs[2].set_title("Histogram of Sample Means for Binomial Distribution")
axs[2].set_xlabel("Sample Mean")
axs[2].set_ylabel("Density")

fig.tight_layout()
plt.show()
```

# Case study for central limit theoram

```python
df=pd.read_csv("titanic.csv")
```

```python
df["Fare"].plot(kind='kde')
```

```python
samples=[]

for i in range(100):
     samples.append(df["Fare"].dropna().sample(50).values.tolist())
```

```python
samples=np.array(samples)
```

```python
sampling_means=samples.mean(axis=1)
```

```python
import seaborn as sns

sns.kdeplot(sampling_means)
```

```python
upper_range=sampling_means.mean()+2*(sampling_means.std()/np.sqrt(50))
lower_range=sampling_means.mean()-2*(sampling_means.std()/np.sqrt(50))

print(f"Upper range: {upper_range} and lower range: {lower_range}")
```
